<p style="font-size:16px;"> Maximum Square Exercise: </p>
<p style="font-size:14px;"> In this exercise, a number, representing the number of lines, is taken from the user. Then, the program calculates how many smaller squares can be formed within a larger square using that number of lines. Finally, the program outputs the maximum number of squares that can be created based on the number of lines entered by the user.</p>

<p style="font-size:16px;"> 
Input: 3
</p>
<p style="font-size:16px;">
Output: 6 
</p>

<p style="font-size:16px;"> 
Input: 4
</p>
<p style="font-size:16px;">
Output: 10 
</p>

```python
n = int(input(""))
print(max((i+1)*(j+1) for i in range(n) for j in range(n-1, -1, -1)
          if i + j == n))
```
