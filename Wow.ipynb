<p style="font-size:14px;">In this project, the user enters a number, and the rpogram adds that many "o" characters to the word "Wow!". 
For example:</p>

<p style="font-size:14px;"> 
Input  : 1
</p>
<p style="font-size:14px;">
Output : Wow!
</p>

<p style="font-size:14px;"> 
Input  : 2
</p>
<p style="font-size:14px;"> 
Output : Woow!
</p>


```python
n = int(input(""))
wow = 'Wow!'
x = ''
for i in range(1, n):
    x += 'o'
wow = wow[0] + x + wow[1:]
print(wow)
```
